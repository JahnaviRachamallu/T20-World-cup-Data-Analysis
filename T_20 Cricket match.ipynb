## Batting Details

```python
# location of the dataset
with open('t20_wc_batting_summary.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['battingSummary'])
batting = pd.DataFrame(all_records)
batting.tail()
```

```python
# adding a new column - out/not_out based on length of value in dismissal column(if len is >0 then it is out if not not_out)

batting["Out/not_Out"]=batting.dismissal.apply(lambda x: "Out" if len(x)>0 else "not_Out")
batting.head(20)
```

```python
# last 5 records
batting.tail()
```

```python
 # dropping the column - dismissal
batting.drop(columns=["dismissal"],inplace=True)  
batting.head()
```

```python
# removing the special charaters
batting["batsmanName"]=batting["batsmanName"].apply(lambda x: x.replace("â€",""))
batting["batsmanName"]=batting["batsmanName"].apply(lambda x: x.replace("\xa0",""))
batting.head(10)
```

```python
match.head()
```

```python
# Creating a dictionary with key of columns -team1 and team2 
# values-match_id

match_dictionary = {}
for index,row in match.iterrows():
    key1 = row["team1"] + " Vs " + row["team2"]
    key2 = row["team2"] + " Vs " + row["team1"]
    match_dictionary[key1] = row["match_id"]
    match_dictionary[key2] = row["match_id"]

match_dictionary
```

```python
batting["match_id"] = batting["match"].map(match_dictionary)

batting.head()
```

## converting to csv file

```python
batting.to_csv("CSV Files/batting.csv", index = False )

```
